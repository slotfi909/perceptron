## functions

```python
def logistic_function(t):
    t=np.float128(t)
    return 1.0 / (1 + np.exp(-t))
```

```python
def cost_function(theta,x,y,intercept):
    nan = 0 * float("-inf")
    n = len(y)
    y_pred = logistic_function(np.dot(x , theta)+intercept).reshape(n,1)
    error_first_param = y * np.log(y_pred)
    error_second_param = (1 - y) * np.log(1 - y_pred)
    
    # python treats "0 * inf" as "nan", even if the zero is absolute, so we
    # have to manually set all of "0 * inf"s to 0.
    for i in range(len(error_first_param)):

        if math.isnan(error_first_param[i][0]):
            error_first_param[i] = [0]

        if math.isnan(error_second_param[i][0]):
            error_second_param[i] = [0]

    error = error_first_param + error_second_param
    cost = -1 / n * sum(error)
    return cost
```

```python
def calc_derative_theta(x,y,theta,intercept):
    temp_list=[]
    n_samples,n_features=x.shape
    for j in range(n_features): 
        sum=0
        for i in range(n_samples):
            sum+=(logistic_function(x[i].dot(theta)+intercept[i]) - y[i])*x[i][j]
        sum/=n_samples
        temp_list.append(np.array(sum))
    return np.array(temp_list)
```

```python
def calc_derative_intercept(x,y,theta,intercept): 
    temp_list=[]
    n_samples,n_features=x.shape
    sum=0
    for i in range(n_samples): 
         sum+=(logistic_function(x[i].dot(theta)+intercept[i]) - y[i])
    sum/=n_samples
    temp_list.append(np.array(sum))
    return np.array(temp_list)
```

```python
def gradient_descent(alpha,theta,dtheta,intercept,dintercept):
    theta-=alpha*dtheta
    intercept-=alpha*dintercept
    return intercept,theta
```

```python
def random_init(m):
    temp_list=[]
    for i in range(m):
        temp_list.append([random.uniform(-0.03,0.02)])
    return np.array(temp_list)


#for test
my_array=random_init(5)
my_array
```

```python
def logistic_regression(x,y,alpha,num_iterations,add_intercept):
    n_samples,n_features=x.shape
    if(add_intercept):
        intercept=random_init(n_samples)
    else:
        intercept=np.zeros(n_samples).reshape(n_samples,1)
        dintercept=np.zeros(n_samples).reshape(n_samples,1)
        
    print(f'n={n_samples},m={n_features}')
    theta=random_init(n_features)
    
    #start iteration
    for i in range(num_iterations):
        print(f'iteration: {i+1}')
        
        #we need to implement a dynamic alpha(learning) in order to avoid
        #oscillation so we multiply it by 0.9 in each iteration
        alpha*=0.9
        
        x_theta=np.dot(x,theta)
        y_hat=x_theta+intercept
        sigmoid_y_hat=logistic_function(y_hat).reshape(n_samples,1)
        dtheta=calc_derative_theta(x,y,theta,intercept)
        if(add_intercept):
            dintercept=calc_derative_intercept(x,y,theta,intercept)
            
        print(f'cost={cost_function(theta,x,y,intercept)}')
        intercept,theta=gradient_descent(alpha,theta,dtheta,intercept,dintercept)
    
    #end iteration
    print(f'final cost={cost_function(theta,x,y,intercept)}')
    return intercept,theta
    
```
